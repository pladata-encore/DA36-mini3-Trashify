```python
import cv2
import os
```

```python
# image_paths에는 directory에 담겨 있는 모든 파일의 경로정보가 담겨 있음
images_paths = []
directory = 'C:\\Users\\playdata2\\Desktop\\chair_all'
for root, _, files in os.walk(directory):
    for file in files:
          images_paths.append(os.path.join(root, file))
```

```python
save_dir = 'C:\\Users\\playdata2\\Desktop\\resized_data_all' # resize된 데이터를 저장할 디렉토리 경로
target_size = (224, 224)
if 'diningtable' in directory:
    label = 'dining_table'
elif 'cabinet' in directory:
    label = 'cabinet'
elif 'sofa' in directory:
    label = 'sofa'
elif 'chair' in directory:
    label = 'chair'
else:
    raise ValueError('디렉토리명에서 라벨값을 찾을 수 없습니다.')

for i, (image_path) in enumerate(images_paths):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # 리사이즈
    image = cv2.resize(image, target_size)

    label_dir = os.path.join(save_dir, label)
    os.makedirs(label_dir, exist_ok=True)
    save_path = os.path.join(label_dir, f'resized_{i}.jpg')
    cv2.imwrite(save_path, cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
    print(f'Saved resized image to: {save_path}')
```
